```python
from konlpy.tag import Okt

Okt = Okt()
# 예시 문장
sentence = "돼지고기 목살을 맛있께 목살을 굽는다."
```

```python
Okt.morphs(sentence)
```

```python
print(Okt.nouns(sentence))
```

```python
def find_ingredients_in_db(word_list):
    conn = sqlite3.connect("ingredients.db")
    cursor = conn.cursor()

    # 단어 리스트를 튜플로 변환하여 SQL IN 조건 사용
    placeholders = ','.join(['?'] * len(word_list))
    query = f"SELECT name FROM ingredients WHERE name IN ({placeholders})"

    cursor.execute(query, word_list)
    results = cursor.fetchall()

    conn.close()

    # 결과를 리스트 형태로 반환
    return [row[0] for row in results]

# 예제 실행
extracted_words = Okt.nouns(sentence)
matched_ingredients = find_ingredients_in_db(extracted_words)
print("🔍 DB에 존재하는 재료:", matched_ingredients)
```

```python
def print_tree(parent_id=None, level=0):
    conn = sqlite3.connect("ingredients.db")
    cursor = conn.cursor()

    cursor.execute("""
    SELECT id, name FROM ingredients WHERE parent_id IS ?
    """, (parent_id,))

    results = cursor.fetchall()
    conn.close()

    for row in results:
        print(" " * (level * 4) + f"- {row[1]}")
        print_tree(row[0], level + 1)

print_tree()  # 트리 형태로 전체 출력
```

이 단계에서 한 부분: 특정 문장에서 해당 단어를 추출 해 올 수 있게 되었다.
이제, 레시피가 있을 때, 해당 레시피에서 재료를 추출할 수 있게 되었다.
